#### Necessary imports

```python
from pathlib import Path
import os, json

from arcgis.learn import prepare_data, ImageCaptioner
from arcgis.gis import GIS
```

#### Model training

```python
filepath = r'D:\Python Projects\Image Captioning\ImageCaptioning'
```

```python
data_path = Path(os.path.join(os.path.splitext(filepath)[0]))
```

```python
data = prepare_data(data_path, 
                    chip_size=224,
                    batch_size=4,
                    dataset_type='ImageCaptioning')
```

#### Visualize Training Data

```python
data.show_batch()
```

#### Load Model Architecture

```python
ic = ImageCaptioner(data, backbone='resnet50')
```

```python
lr = ic.lr_find()
```

```python
lr 
```

#### Train the Model

```python
ic.fit(1, lr, early_stopping=True)
```

#### Visualize results on validation set

```python
ic.show_results()
```

#### Evaluate Model Performance

```python
ic.bleu_score()
```

#### Save the Model

```python
ic.save('image-captioner-01epochs')
```

#### Predication on test image

```python
ic.predict(r'\image-captioner\test_img.tif')
```
